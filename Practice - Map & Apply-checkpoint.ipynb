```python
import pandas as pd
movies = pd.read_csv("http://bit.ly/kaggletrain")
#movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()

```

```python
movies['sex_num'] = movies.Sex.map({'male':1,'female':2})
movies.loc[0:5 , ['sex_num','Sex']]
```

## Using Map method to create a new column for Gender in a DataFrame

```python
movies['Gender'] = movies.Sex.map({'male':'M','female':'F'})
movies.loc[:,['Sex','Gender']]
```
